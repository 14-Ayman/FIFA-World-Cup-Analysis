```python
# Step 1: Importing Libraries and Setting Up the Environment

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="pastel")

pd.set_option('display.max_columns', None)
pd.set_option('display.precision', 2)

%matplotlib inline

print("Libraries imported and environment ready.")
```

```python
# Step 2: Loading and Previewing the FIFA Datasets

# Read the CSV files into Pandas DataFrames
cups = pd.read_csv(r"C:\Users\ayman\I\FIFA WC data\WorldCups.csv")
matches = pd.read_csv(r"C:\Users\ayman\I\FIFA WC data\WorldCupMatches.csv")
players = pd.read_csv(r"C:\Users\ayman\I\FIFA WC data\WorldCupPlayers.csv")

# Confirm successful load by printing their shapes (rows, columns)
print("Data loaded successfully\n")
print(f"WorldCups       → {cups.shape[0]} rows × {cups.shape[1]} columns")
print(f"WorldCupMatches → {matches.shape[0]} rows × {matches.shape[1]} columns")
print(f"WorldCupPlayers → {players.shape[0]} rows × {players.shape[1]} columns")

# Preview the first few rows of each dataset
print("\n Preview of WorldCups dataset:")
display(cups.head())

print("\nPreview of WorldCupMatches dataset:")
display(matches.head())

print("\nPreview of WorldCupPlayers dataset:")
display(players.head())

# Basic info and column details
print("\nBasic Info:")
print("\nWorldCups ")
cups.info()
print("\nWorldCupMatches")
matches.info()
print("\nWorldCupPlayers")
players.info()
```

```python
# STEP 3 — DATA CLEANING 

import pandas as pd
import os

save_path = r"C:\Users\ayman\Project 2"

os.makedirs(save_path, exist_ok=True)
os.makedirs(f"{save_path}/cleaned_data", exist_ok=True)

# ----- Missing values -----
print(" Missing Values Before Cleaning:\n")
print("WorldCups:\n", cups.isnull().sum(), "\n")
print("WorldCupMatches:\n", matches.isnull().sum(), "\n")
print("WorldCupPlayers:\n", players.isnull().sum(), "\n")

# ----- Drop duplicates -----
cups.drop_duplicates(inplace=True)
matches.drop_duplicates(inplace=True)
players.drop_duplicates(inplace=True)

# ----- Fix Year column -----
cups['Year'] = pd.to_numeric(cups['Year'], errors='coerce')
cups = cups.dropna(subset=['Year'])
cups['Year'] = cups['Year'].astype(int)

matches['Year'] = pd.to_numeric(matches['Year'], errors='coerce')
matches = matches.dropna(subset=['Year'])
matches['Year'] = matches['Year'].astype(int)

# ----- Clean string columns -----
string_cols = matches.select_dtypes(include='object').columns
for col in string_cols:
    matches[col] = matches[col].astype(str).str.strip()

# ----- Fill missing numeric values -----
numeric_cols = matches.select_dtypes(include=['float64', 'int64']).columns
matches[numeric_cols] = matches[numeric_cols].fillna(0)

# ----- Save cleaned files -----
cups.to_csv(f"{save_path}/cleaned_data/WorldCups_clean.csv", index=False)
matches.to_csv(f"{save_path}/cleaned_data/WorldCupMatches_clean.csv", index=False)
players.to_csv(f"{save_path}/cleaned_data/WorldCupPlayers_clean.csv", index=False)

print("\n Step 3 Completed Successfully!")
print(f" Cleaned data saved in: {save_path}/cleaned_data")
```

```python
# STEP 4 — EDA ( EXPLORATORY DATA ANALYSIS )

import os
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set(style="whitegrid")


save_path = r"C:\Users\ayman\Project 2"
os.makedirs(f"{save_path}/results", exist_ok=True)

print("Starting Advanced Step 4: Exploratory Data Analysis...\n")

sns.set(style="whitegrid")

# --------------- 4.1 WINNERS BAR CHART ----------------
winners = cups['Winner'].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=winners.index, y=winners.values)
plt.xticks(rotation=45)
plt.title("Most Successful FIFA World Cup Nations")
plt.tight_layout()
plt.savefig(f"{save_path}/results/01_top_winning_countries.png", dpi=300)
plt.show()

# --------------- 4.2 GOALS PER TOURNAMENT ----------------
plt.figure(figsize=(10,5))
sns.lineplot(x="Year", y="GoalsScored", data=cups, marker="o")
plt.title("Total Goals per World Cup")
plt.tight_layout()
plt.savefig(f"{save_path}/results/02_goals_per_tournament.png", dpi=300)
plt.show()

# --------------- 4.3 AVERAGE GOALS PER MATCH ----------------
matches['Total Goals'] = matches['Home Team Goals'] + matches['Away Team Goals']
avg_goals = matches.groupby("Year")["Total Goals"].mean().reset_index()

plt.figure(figsize=(10,5))
sns.lineplot(x="Year", y="Total Goals", data=avg_goals, marker="o")
plt.title("Average Goals per Match Over the Years")
plt.tight_layout()
plt.savefig(f"{save_path}/results/03_average_goals_per_match.png", dpi=300)
plt.show()

# --------------- 4.4 ATTENDANCE OVER YEARS ----------------
plt.figure(figsize=(10,5))
sns.lineplot(x="Year", y="Attendance", data=cups, marker="o")
plt.gca().invert_yaxis()  
plt.title("World Cup Attendance (Inverted View)")
plt.tight_layout()
plt.savefig(f"{save_path}/results/04_attendance_inverted_yaxis.png", dpi=300)
plt.show()


# --------------- 4.5 HOST NATION PERFORMANCE ----------------
cups['Host Won?'] = cups['Winner'] == cups['Country']
host_summary = cups['Host Won?'].value_counts()

plt.figure(figsize=(6,6))
host_summary.plot(kind="pie", autopct="%1.1f%%", labels=["Didn't Win", "Won"])
plt.title("How Often Did the Host Nation Win?")
plt.tight_layout()
plt.savefig(f"{save_path}/results/05_host_advantage.png", dpi=300)
plt.show()

# --------------- 4.6 MOST USED STADIUM ----------------
stadium_counts = matches['Stadium'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(y=stadium_counts.index, x=stadium_counts.values)
plt.title("Top 10 Most Used Stadiums in World Cup")
plt.tight_layout()
plt.savefig(f"{save_path}/results/06_most_used_stadiums.png", dpi=300)
plt.show()

# --------------- 4.7 MOST USED CITIES ----------------
city_counts = matches['City'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(y=city_counts.index, x=city_counts.values)
plt.title("Top 10 Cities With Most World Cup Matches")
plt.tight_layout()
plt.savefig(f"{save_path}/results/07_top_cities.png", dpi=300)
plt.show()

# --------------- 4.8 GOAL DIFFERENCE DISTRIBUTION ----------------
matches['Goal Difference'] = abs(matches['Home Team Goals'] - matches['Away Team Goals'])

plt.figure(figsize=(10,5))
sns.histplot(matches['Goal Difference'], bins=10, kde=True)
plt.title("Distribution of Goal Differences in Matches")
plt.tight_layout()
plt.savefig(f"{save_path}/results/08_goal_difference_distribution.png", dpi=300)
plt.show()

# --------------- 4.9 MOST COMMON MATCH STAGES ----------------
stage_counts = matches['Stage'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(y=stage_counts.index, x=stage_counts.values)
plt.title("Most Frequently Played World Cup Stages")
plt.tight_layout()
plt.savefig(f"{save_path}/results/09_top_stages.png", dpi=300)
plt.show()

# --------------- 4.10 TEAMS WITH MOST APPEARANCES ----------------
teams = pd.concat([matches['Home Team Name'], matches['Away Team Name']])
team_counts = teams.value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(y=team_counts.index, x=team_counts.values)
plt.title("Teams With the Most World Cup Match Appearances")
plt.tight_layout()
plt.savefig(f"{save_path}/results/10_top_match_appearances.png", dpi=300)
plt.show()

# --------------- 4.11 CORRELATION HEATMAP ----------------
plt.figure(figsize=(8,6))
sns.heatmap(matches.select_dtypes(include=['float64','int64']).corr(), cmap="coolwarm")
plt.title("Match Statistics Correlation Heatmap")
plt.tight_layout()
plt.savefig(f"{save_path}/results/11_correlation_heatmap.png", dpi=300)
plt.show()

```

```python
# STEP 5 — INSIGHTS + FINAL REPORT 

import os
import pandas as pd

save_path = r"C:\Users\ayman\Project 2"
os.makedirs(f"{save_path}/reports", exist_ok=True)

insights = []

# ----- INSIGHT 1: Most successful team -----
top_team = cups['Winner'].value_counts().idxmax()
wins = cups['Winner'].value_counts().max()
insights.append(f"The most successful team in FIFA World Cup history is {top_team} with {wins} titles.")

# ----- INSIGHT 2: Most hosted country -----
top_host = cups['Country'].value_counts().idxmax()
host_count = cups['Country'].value_counts().max()
insights.append(f" The country that has hosted the most World Cups is {top_host}, hosting {host_count} times.")

# ----- INSIGHT 3: Highest scoring world cup -----
max_goals_row = cups.loc[cups['GoalsScored'].idxmax()]
insights.append(
    f" The highest scoring World Cup was {int(max_goals_row['Year'])} "
    f"with {int(max_goals_row['GoalsScored'])} total goals."
)

# ----- INSIGHT 4: Lowest scoring world cup -----
min_goals_row = cups.loc[cups['GoalsScored'].idxmin()]
insights.append(
    f"The lowest scoring World Cup was {int(min_goals_row['Year'])} "
    f"with only {int(min_goals_row['GoalsScored'])} goals."
)

# ----- INSIGHT 5: Host advantage -----
host_wins = cups[cups['Winner'] == cups['Country']]
host_rate = (host_wins.shape[0] / cups.shape[0]) * 100
insights.append(
    f"Host nations have won {host_wins.shape[0]} tournaments "
    f"({host_rate:.1f}% host win rate)."
)

# ----- INSIGHT 6: Avg goals per match trend -----
matches['Total Goals'] = matches['Home Team Goals'] + matches['Away Team Goals']
avg_goals_per_year = matches.groupby('Year')['Total Goals'].mean()

trend = "decreased" if avg_goals_per_year.iloc[-1] < avg_goals_per_year.iloc[0] else "increased"

insights.append(
    f" Average goals per match have {trend} from "
    f"{avg_goals_per_year.iloc[0]:.2f} to {avg_goals_per_year.iloc[-1]:.2f}."
)

# ----- INSIGHT 7: Highest attendance -----
max_attendance = cups.loc[cups['Attendance'].idxmax()]
insights.append(
    f" The highest-attended World Cup was {int(max_attendance['Year'])} "
    f"with a total attendance of {int(max_attendance['Attendance']):,} spectators."
)

# ----- INSIGHT 8: Most used stadium -----
top_stadium = matches['Stadium'].value_counts().idxmax()
stadium_matches = matches['Stadium'].value_counts().max()
insights.append(
    f" The most used stadium in World Cup history is {top_stadium} "
    f"with {stadium_matches} matches hosted."
)

# ----- INSIGHT 9: Most used city -----
top_city = matches['City'].value_counts().idxmax()
city_matches = matches['City'].value_counts().max()
insights.append(
    f"The city that hosted the highest number of World Cup matches is "
    f"{top_city} with {city_matches} games."
)

# ----- INSIGHT 10: Largest win margin -----
matches['Goal Difference'] = abs(matches['Home Team Goals'] - matches['Away Team Goals'])
max_margin = matches['Goal Difference'].max()
insights.append(
    f" The biggest winning margin recorded in a World Cup match is {int(max_margin)} goals."
)

# ----- INSIGHT 11: Most common stage played -----
top_stage = matches['Stage'].value_counts().idxmax()
insights.append(f" The most frequently played stage in World Cup matches is the '{top_stage}'.")

# ----- INSIGHT 12: Teams with most appearances -----
teams = pd.concat([matches['Home Team Name'], matches['Away Team Name']])
team_counts = teams.value_counts()
most_played_team = team_counts.idxmax()
most_played_count = team_counts.max()
insights.append(
    f"🇺🇳 The team with the most match appearances is {most_played_team} "
    f"with {most_played_count} total matches played."
)

# ----- SAVE INSIGHTS -----
with open(f"{save_path}/reports/final_insights.txt", "w", encoding="utf-8") as f:
    for line in insights:
        f.write(line + "\n")

print(" Expanded insights saved to final_insights.txt")

# ----- SAVE SUMMARY REPORT -----
markdown_report = f"""
# FIFA World Cup Analysis — Final Report

##  Key Insights
{chr(10).join(insights)}

---

##  Visuals
Graphs are saved here:  
 {save_path}/results

##  Cleaned Data
Saved in:  
{save_path}/cleaned_data  
and  
 {save_path}/reports

---

### Project by Ayman Abbas
"""

with open(f"{save_path}/reports/summary_report.md", "w", encoding="utf-8") as f:
    f.write(markdown_report)

print(" Updated summary_report.md saved.")
print("\n Step 5 Completed Successfully!")
```
